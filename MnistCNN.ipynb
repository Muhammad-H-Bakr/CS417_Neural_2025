```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras  # type: ignore
from tensorflow.keras import layers  # type: ignore
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# -------------------------------------------------------------------
# 1. Load Dataset
# -------------------------------------------------------------------
(X_train_full, y_train_full), (X_test, y_test) = tf.keras.datasets.mnist.load_data()  # type: ignore

print("Dataset loaded successfully!")
print(f"Training data shape: {X_train_full.shape}")
print(f"Training labels shape: {y_train_full.shape}")
print(f"Test data shape: {X_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# -------------------------------------------------------------------
# 2. Train/Validation/Test Split
# -------------------------------------------------------------------
# Split the training data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.1, random_state=42
)

print(f"\nAfter split:")
print(f"X_train shape: {X_train.shape}")
print(f"X_val shape: {X_val.shape}")
print(f"X_test shape: {X_test.shape}")

# -------------------------------------------------------------------
# 3. Data Preprocessing
# -------------------------------------------------------------------
# Normalize pixel values to [0, 1] and add channel dimension for CNN
X_train = X_train.astype("float32") / 255.0
X_val = X_val.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0

# Reshape data to include channel dimension (28, 28, 1) for grayscale
X_train = X_train.reshape(-1, 28, 28, 1)
X_val = X_val.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

print(f"\nAfter preprocessing:")
print(f"X_train shape: {X_train.shape}")
print(f"X_val shape: {X_val.shape}")
print(f"X_test shape: {X_test.shape}")

# Convert labels to categorical one-hot encoding
y_train_categorical = keras.utils.to_categorical(y_train, 10)
y_val_categorical = keras.utils.to_categorical(y_val, 10)
y_test_categorical = keras.utils.to_categorical(y_test, 10)

print(f"\nLabel shapes after one-hot encoding:")
print(f"y_train_categorical shape: {y_train_categorical.shape}")
print(f"y_val_categorical shape: {y_val_categorical.shape}")

# -------------------------------------------------------------------
# 4. CNN Model Architecture
# -------------------------------------------------------------------
model = keras.Sequential(
    [
        # Input layer
        keras.Input(shape=(28, 28, 1)),
        # First Convolutional Block
        layers.Conv2D(32, (3, 3), activation="relu", kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation="relu", kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        # Second Convolutional Block
        layers.Conv2D(64, (3, 3), activation="relu", kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation="relu", kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        # Third Convolutional Block
        layers.Conv2D(128, (3, 3), activation="relu", kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.Dropout(0.25),
        # Classifier
        layers.Flatten(),
        layers.Dense(512, activation="relu", kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation="relu", kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(10, activation="softmax"),  # 10 classes for digits 0-9
    ]
)

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",  # Use categorical crossentropy for multi-class
    metrics=["accuracy", "categorical_accuracy"],  # Track accuracy metrics
)

# Display model architecture
print("\n" + "=" * 50)
print("MODEL ARCHITECTURE")
print("=" * 50)
model.summary()

# -------------------------------------------------------------------
# 5. Callbacks
# -------------------------------------------------------------------
early_stopping = keras.callbacks.EarlyStopping(
    monitor="val_accuracy", patience=15, restore_best_weights=True, verbose=1
)

lr_scheduler = keras.callbacks.ReduceLROnPlateau(
    monitor="val_accuracy", factor=0.5, patience=8, min_lr=1e-7, verbose=1
)

# Optional: Model checkpoint to save best model
checkpoint = keras.callbacks.ModelCheckpoint(
    "best_mnist_model.keras", monitor="val_accuracy", save_best_only=True, verbose=0
)

# -------------------------------------------------------------------
# 6. Training
# -------------------------------------------------------------------
print("\n" + "=" * 50)
print("TRAINING STARTED")
print("=" * 50)

history = model.fit(
    X_train,
    y_train_categorical,
    epochs=100,
    batch_size=128,
    validation_data=(X_val, y_val_categorical),
    callbacks=[early_stopping, lr_scheduler, checkpoint],
    verbose=1,
)

# -------------------------------------------------------------------
# 7. Evaluation
# -------------------------------------------------------------------
print("\n" + "=" * 50)
print("FINAL EVALUATION")
print("=" * 50)

# Evaluate on test set
test_loss, test_accuracy, test_cat_accuracy = model.evaluate(
    X_test, y_test_categorical, verbose=0
)

# Evaluate on validation set for comparison
val_loss, val_accuracy, val_cat_accuracy = model.evaluate(
    X_val, y_val_categorical, verbose=0
)

print(f"Validation Accuracy : {val_accuracy:.4f}")
print(f"Test Accuracy       : {test_accuracy:.4f}")
print(f"Test Loss           : {test_loss:.4f}")

# -------------------------------------------------------------------
# 8. Predictions and Analysis
# -------------------------------------------------------------------
print("\n" + "=" * 50)
print("PREDICTION ANALYSIS")
print("=" * 50)

# Get predictions
y_pred_proba = model.predict(X_test)
y_pred = np.argmax(y_pred_proba, axis=1)

# Show some predictions vs actual
print("\nFirst 15 predictions vs actual:")
print("Predicted | Actual | Correct")
print("-" * 25)

correct_count = 0
for i in range(15):
    pred = y_pred[i]
    actual = y_test[i]
    correct = "✓" if pred == actual else "✗"
    if correct == "✓":
        correct_count += 1
    print(f"    {pred}     |    {actual}   |   {correct}")

print(f"\nAccuracy on first 15 samples: {correct_count}/15 = {correct_count/15:.2%}")

# -------------------------------------------------------------------
# 9. Detailed Performance Analysis
# -------------------------------------------------------------------
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

print("\n" + "=" * 50)
print("DETAILED PERFORMANCE METRICS")
print("=" * 50)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# -------------------------------------------------------------------
# 10. Training History Visualization
# -------------------------------------------------------------------
plt.figure(figsize=(15, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# -------------------------------------------------------------------
# 11. Show some test images with predictions
# -------------------------------------------------------------------
print("\n" + "=" * 50)
print("VISUAL PREDICTION EXAMPLES")
print("=" * 50)

fig, axes = plt.subplots(3, 5, figsize=(15, 9))
fig.suptitle("Test Images with Predictions", fontsize=16)

for i in range(15):
    row = i // 5
    col = i % 5

    # Display image
    axes[row, col].imshow(X_test[i].reshape(28, 28), cmap="gray")

    # Set title with prediction result
    pred = y_pred[i]
    actual = y_test[i]
    color = "green" if pred == actual else "red"
    axes[row, col].set_title(f"Pred: {pred} | Actual: {actual}", color=color)
    axes[row, col].axis("off")

plt.tight_layout()
plt.show()

print(f"\n🎯 FINAL TEST ACCURACY: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras # type: ignore
import matplotlib.pyplot as plt

# -------------------------------------------------------------------
# 1. Load the Saved Model
# -------------------------------------------------------------------
print("Loading saved model...")
model = keras.models.load_model('best_mnist_model.keras')
print("✅ Model loaded successfully!")

# -------------------------------------------------------------------
# 2. Load Fresh Test Data
# -------------------------------------------------------------------
print("\nLoading MNIST test data...")
(_, _), (X_test, y_test) = tf.keras.datasets.mnist.load_data() # type: ignore

# Preprocess the test data (same as during training)
X_test = X_test.astype('float32') / 255.0
X_test = X_test.reshape(-1, 28, 28, 1)  # Add channel dimension
y_test_categorical = keras.utils.to_categorical(y_test, 10)

print(f"Test data shape: {X_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# -------------------------------------------------------------------
# 3. Evaluate the Model
# -------------------------------------------------------------------
print("\n" + "="*50)
print("EVALUATING SAVED MODEL")
print("="*50)

# Evaluate on test set
test_loss, test_accuracy, test_cat_accuracy = model.evaluate(
    X_test, y_test_categorical, verbose=1
)

print(f"\n📊 Model Performance:")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

# -------------------------------------------------------------------
# 4. Make Predictions
# -------------------------------------------------------------------
print("\n" + "="*50)
print("MAKING PREDICTIONS")
print("="*50)

# Get predictions
y_pred_proba = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)

# -------------------------------------------------------------------
# 5. Test on Specific Examples
# -------------------------------------------------------------------
def test_single_image(model, image_array, true_label=None):
    """
    Test the model on a single image
    """
    # Ensure proper shape (1, 28, 28, 1)
    if image_array.ndim == 2:
        image_array = image_array.reshape(1, 28, 28, 1)
    elif image_array.ndim == 3:
        image_array = image_array.reshape(1, 28, 28, 1)

    # Make prediction
    prediction = model.predict(image_array, verbose=0)
    predicted_class = np.argmax(prediction[0])
    confidence = np.max(prediction[0])

    # Display results
    plt.imshow(image_array[0, :, :, 0], cmap='gray')
    if true_label is not None:
        plt.title(f'True: {true_label} | Pred: {predicted_class} (conf: {confidence:.3f})',
                 color='green' if predicted_class == true_label else 'red')
    else:
        plt.title(f'Predicted: {predicted_class} (conf: {confidence:.3f})')
    plt.axis('off')
    plt.show()

    # Show top 3 predictions
    top_3_indices = np.argsort(prediction[0])[-3:][::-1]
    top_3_confidences = prediction[0][top_3_indices]

    print(f"Top 3 predictions:")
    for i, (cls, conf) in enumerate(zip(top_3_indices, top_3_confidences)):
        print(f"  {i+1}. Digit {cls}: {conf:.4f}")

    return predicted_class, confidence

# -------------------------------------------------------------------
# 6. Test Random Samples
# -------------------------------------------------------------------
print("\n" + "="*50)
print("TESTING RANDOM SAMPLES FROM TEST SET")
print("="*50)

# Test 5 random samples
np.random.seed(42)  # For reproducible results
random_indices = np.random.choice(len(X_test), 5, replace=False)

print("Testing 5 random samples:")
for i, idx in enumerate(random_indices):
    print(f"\nSample {i+1}:")
    true_label = y_test[idx]
    pred, conf = test_single_image(model, X_test[idx], true_label)

    if pred == true_label:
        print(f"✅ CORRECT! Model predicted {pred} with {conf:.3f} confidence")
    else:
        print(f"❌ WRONG! Model predicted {pred} but true label is {true_label}")

# -------------------------------------------------------------------
# 7. Test Specific Digits (0-9)
# -------------------------------------------------------------------
print("\n" + "="*50)
print("TESTING ONE EXAMPLE OF EACH DIGIT")
print("="*50)

# Find first occurrence of each digit in test set
for digit in range(10):
    indices = np.where(y_test == digit)[0]
    if len(indices) > 0:
        idx = indices[0]
        true_label = y_test[idx]
        pred, conf = test_single_image(model, X_test[idx], true_label)

        status = "✅ CORRECT" if pred == true_label else "❌ WRONG"
        print(f"Digit {digit}: {status} (confidence: {conf:.3f})")

# -------------------------------------------------------------------
# 8. Analyze Misclassifications
# -------------------------------------------------------------------
print("\n" + "="*50)
print("MISCLASSIFICATION ANALYSIS")
print("="*50)

misclassified_indices = np.where(y_pred != y_test)[0]
print(f"Total misclassified: {len(misclassified_indices)}/{len(y_test)}")
print(f"Error rate: {len(misclassified_indices)/len(y_test)*100:.2f}%")

if len(misclassified_indices) > 0:
    print(f"\nFirst 5 misclassified examples:")
    for i in range(min(5, len(misclassified_indices))):
        idx = misclassified_indices[i]
        pred = y_pred[idx]
        actual = y_test[idx]
        confidence = np.max(y_pred_proba[idx])
        print(f"  Sample {idx}: Predicted {pred} | Actual {actual} | Confidence: {confidence:.3f}")

# -------------------------------------------------------------------
# 9. Test Your Own Handwritten Digits (Optional)
# -------------------------------------------------------------------
print("\n" + "="*50)
print("TESTING WITH CUSTOM IMAGES")
print("="*50)
print("To test your own handwritten digits:")
print("1. Create a 28x28 grayscale image with white digit on black background")
print("2. Save it as a PNG file")
print("3. Use the function below to test it")

def test_custom_image(model, image_path):
    """
    Test the model on a custom image file
    """
    from PIL import Image

    # Load and preprocess image
    img = Image.open(image_path).convert('L')  # Convert to grayscale
    img = img.resize((28, 28))  # Resize to 28x28
    img_array = np.array(img)

    # Invert if needed (MNIST has white digits on black background)
    if np.mean(img_array) > 127:  # If background is light
        img_array = 255 - img_array

    # Normalize
    img_array = img_array.astype('float32') / 255.0

    return test_single_image(model, img_array)

# Example usage for custom images (uncomment and modify path):
# test_custom_image(model, 'my_digit_7.png')

# -------------------------------------------------------------------
# 10. Final Summary
# -------------------------------------------------------------------
print("\n" + "="*60)
print("🎯 LOADED MODEL PERFORMANCE SUMMARY")
print("="*60)
print(f"🏆 Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"📊 Test Loss: {test_loss:.4f}")
print(f"🔍 Misclassified: {len(misclassified_indices)}/10000 samples")
print(f"🎨 Model: best_mnist_model.keras")
print("="*60)
print("✅ Model is ready for use!")
```
